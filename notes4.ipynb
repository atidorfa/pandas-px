```python
people = {
    "first": ["Gonzalo", "Jane", "Joe"],
    "last": ["Rivero Baud", "Doe", "Doe"],
    "email": ["burrito@gmail.com", 'janedoe@email.com', "johndoe@email.com"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['last'] == 'Doe'
```

```python
filt = df['last'] == 'Doe'
```

```python
filt
```

```python
df[filt]
```

```python
df[df['last'] == 'Doe']
```

```python
df.loc[filt]
```

```python
df.loc[filt, 'email']
```

```python
filt2 = (df['last'] == 'Doe') & (df['first'] == 'Jane')
```

```python
filt2
```

```python
df.loc[filt2, 'email']
```

```python
filt3 = (df['last'] == 'Rivero Baud') | (df['first'] == 'Jane')
```

```python
filt3
```

```python
df.loc[filt3, 'email']
```

```python
df.loc[~filt3, 'email']
```

```python
df2 = pd.read_csv('data/survey_results_public.csv')
```

```python
df2
```

```python
high_salary_filter = (df2['ConvertedCompYearly'] < 400000)
```

```python
df2.loc[high_salary_filter].head()
```

```python
df2.loc[high_salary_filter, ['Country', 'LanguageHaveWorkedWith', 'ConvertedCompYearly']].sort_values('ConvertedCompYearly', ascending=False).head(70)
```

```python
countries = ['United States', 'India', 'Argentina', 'United Kingdom', 'Germany', 'Canada']
c_filter = df2['Country'].isin(countries)
```

```python
df2.loc[c_filter, 'Country']
```

```python
df2['LanguageHaveWorkedWith']
```

```python
l_filter = df2['LanguageHaveWorkedWith'].str.contains('Python', na=False)
```

```python
df2.loc[l_filter, 'LanguageHaveWorkedWith']
```
